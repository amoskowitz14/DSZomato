# Clustering of Restaraunts

```python
# Convert categorical to dummy variables

df_kmeans = df.drop_duplicates(subset =['name', 'address'])[['online_order','book_table','rest_type','cuisines']]

#for cat in ['online_order','book_table','rest_type','cuisines']:

x1 = pd.get_dummies(df_kmeans["online_order"],prefix='online_order',drop_first=True)
x2 = pd.get_dummies(df_kmeans["book_table"],prefix='book_table',drop_first=True)
x3 = pd.get_dummies(df_kmeans["rest_type"],prefix='rest_type',drop_first=True)
x4 = pd.get_dummies(df_kmeans["cuisines"],prefix='cuisines',drop_first=True)

df_kmeans = x1.merge(x2, left_index = True, right_index = True).merge(x3, left_index = True, right_index = True).merge(x4, left_index = True, right_index = True)
df_kmeans
#df_kmeans = pd.merge([x1,x2], axis=0, ignore_index=True)
```
